```python
# 2. 데이터프레임

## 학생 3명

data = {
    "이름" : ["홍길동", "이순신", "임꺽정"],
    "나이" : [22, 33, 55],
    "학교" : ["서울대", "고려대", "연세대"]
}

student = pd.DataFrame(data)
```

```python
## 3. 데이터에 접근
## 3.1 기본적으로 시리즈 단위로 접근

student['이름']
student['나이'][0]
```

```python
## 3.2 데이터프레임 열을 여러개 선택
df2 = student[['이름', '나이']]

```

```python
## 3.3 데이터 프레임에서 행을 선택하는 방법
selected_row = df2.loc[0]
selected_row
```

```python
## 3.4 행과 열을 동시에 선택
student
selected_data = student.loc[[0, 1], ['이름', '나이']]

selected_data
```

```python
## 0번 2번 학생의 이름과 학교를 선택해주세요.

student.loc[[0,2], ['이름', '학교']]
```

```python
# 5. loc, iloc 이용한 데이터 접근
## 5.1 loc 단일 선택 -> row 기준
df.loc[0]

## 5.2 다중 선택
df.loc[[0,2,5]]

## 5.3 슬라이싱
df2 = df.loc[0:4:2]

## 5.4 논리 인덱싱
df2.loc[[False, False, True]]

## 5.5 row와 column을 같이 선택
df.loc[0, '찾으신금액']
df.loc[0, ['찾으신금액',	'맡기신금액',	'거래후 잔액'	]]
df.loc[[0, 2, 4], ['찾으신금액',	'맡기신금액',	'거래후 잔액'	]]
df.loc[0:4:2   , '찾으신금액' : '거래후 잔액']

## 5.6 특정 조건을 만족하는 행만 선택하고 싶을 때
### 5.6.1 찾으신금액이 100000원 이상인 행만 선택.
df.loc[df['찾으신금액'] > 100000]

### 5.6.2 찾으신금액이 100000원 이상이면서, 적요가 통신요금인것
df.loc[(df['찾으신금액'] > 100000) & (df['적요'] == '통신요금')]
```
