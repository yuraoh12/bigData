```python
import re

def filter_string(s):
    return ''.join(re.findall('[가-힣a-zA-Z0-9\s]', s))

input_str = "안녕하세요! Hello, 12345!@#$%"
filtered_str = filter_string(input_str)
print(filtered_str)  # 출력: 안녕하세요Hello12345
```

```python
from konlpy.tag import Okt

okt = Okt()

okt.morphs(u'단독입찰보다 복수입찰의 경우')


string = "안녕하세요 저는 홍길동입니다. 만나서 반가워요. 저 홍길동은 대전에서 삽니다. 대전은 참 살기 좋아요. 그리고 저 홍길동은 떡볶이를 참 좋아합니다. 대전에 떡볶이집 맛집을 알려드릴게요."

okt.morphs(string)



# 텍스트 클렌징
## 1. stop_word 제거
## 2. 특수 기호 제거
string2 = filter_string(string)

string2

word_list = okt.morphs(string2)

result = []
for word in word_list :
  if word not in stop_words :
    result.append(word)

result
```

```python
word_list = okt.nouns(string2)

result = []
for word in word_list :
  if word not in stop_words :
    result.append(word)

result
```

```python
# 각 단어의 빈도수 구하기

word_count = {}

for word in result :
  if word in word_count :
    word_count[word] += 1
  else :
    word_count[word] = 1


word_count
```

```python
import pandas as pd

kbs = pd.read_json('/content/KBS_news_body.json')
sbs = pd.read_json('/content/SBS_news_body.json')
mbc = pd.read_json('/content/MBC_news_body.json')
```

```python
bodies = pd.concat([kbs, sbs, mbc]).reset_index(drop=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def custom_tokenizer(text):
    text = filter_string(text)
    word_list = okt.morphs(text)
    result = []
    for word in word_list :
      if word not in stop_words :
        result.append(word)

    result
    return result
```

```python
vectorizer = CountVectorizer(tokenizer=custom_tokenizer, max_df=0.9, min_df=3)
X = vectorizer.fit_transform(bodies['body'])

X.shape
```

```python
## 전체 문서에서 너무 많이오는 단어, 너무 적게 나오는 단어
## 최대 빈도수, 최소 빈도수

## 규제, 피처 줄이기
X.shape

```

```python
target = bodies['section']
train = X.toarray()
target
```

```python
from sklearn.model_selection import train_test_split

trd, tsd, trt, tst = train_test_split(train, target, train_size=0.8, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(trd, trt)

```

```python
lr.score(trd, trt), lr.score(tsd, tst)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(lr, tsd, tst, cv=5)

scores
```
