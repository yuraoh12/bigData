```python
import seaborn as sns


sns.set(font = 'NanumBarunGothic',
        rc = {'axes.unicode_minus' : False},
        style="darkgrid")
```

```python
## 선그래프

sns.lineplot(x=[1,2,3,4], y=[2,4,6,8])
```

```python
## 선그래프 -> 시간의 흐름(수치 - 수치)
data_dict = {'data_x': [1, 2, 3, 4, 5], 'data_y': [2, 3, 5, 10, 8]}

df = pd.DataFrame(data_dict)

sns.lineplot(data=df, x='data_x', y='data_y')
plt.xlabel('x축', labelpad=40, fontdict = {'color' : 'b'})
```

```python
## 막대그래프
## 범주형 데이터와 수치간의 관계
sns.barplot(x=['a','b','c','d'], y=[1,2,3,4])
```

```python
## 히스토그램
## 수치와 수치의 관계를 막대그래프
sns.histplot(x=range(1, 101), bins=10)
```

```python
## 산점도
## 두 수치형 변수간의 상관 관계

sns.scatterplot(x=[1,4,7,10], y=[3,11, 5, 8])
```
