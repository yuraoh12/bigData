```python
import pandas as pd
apt1 = pd.read_csv('/content/apt1.csv')
apt2 = pd.read_csv('/content/apt2.csv')
```

```python
tmp1 = apt1.groupby(['지역명', '연도', '월'])['평당분양가격'].mean().reset_index()

tmp2 = pd.melt(apt2, id_vars='지역명', value_vars=apt2.columns.values[1:], var_name='연월', value_name='평당분양가격')

tmp2['연도'] = tmp2['연월'].str.split('년').str[0].astype('int')
tmp2['월'] = tmp2['연월'].str.split('년').str[1].str.replace('월', '').astype('int')

print(tmp1.columns.values)
print(tmp2.columns.values)

tmp3 = tmp2[['지역명', '연도', '월', '평당분양가격']]
```

```python
rst = pd.concat([tmp1, tmp3]).sort_values(['연도', '월', '지역명'])

rst
```
