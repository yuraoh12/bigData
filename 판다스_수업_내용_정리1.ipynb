```python
import pandas as pd

# 이름 리스트
name = [ "Braund, Mr. Owen Harris",
        "Allen, Mr. William Henry",
        "Bonnell, Miss. Elizabeth"]

# 나이 리스트
age = [22, 35, 58]

# 성별 리스트
sex = ["male", "male", "female"]

# 위 리스트를 값으로 가지는 딕셔너리
data = { "Name": name,
    "Age": age,
    "Sex": sex,
}

# 행렬 == 데이터프레임
# 데이터 프레임 만들기
df = pd.DataFrame(data)


# 하나의 열 == 시리즈
df["Age"].max()

ages = df["Age"]

# 시리즈 만들기
d = {'a': 1, 'b': 2, 'c': 3}
ser = pd.Series(data=d, index=['a', 'b', 'c'], name='test')


# 데이터프레임 주요 속성 정보

df.columns # 전체 컬럼 리스트

df.index # 전체 인덱스 리스트

df.index.values # 전체 인덱스 값

df.values # 데이터프레임 전체 배열

df.shape # 행렬의 모양
```

```python
# 데이터 조작

df = pd.DataFrame(data)

# inplace
# api를 사용하면 어떤 연산적용. 적용이 원본에 되느냐, 원본은 유지하고 복사본이 반환되느냐
# False - 복사 반환
# True - 원본 적용

# 컬럼 값을 찾아서 수정. 
df['Sex'].replace({'male' : '남', 'female' : '여'}, inplace=True)

# 컬럼 값의 문자열 수정.
# str.replace 문자열 치환
# str.upper 문자열 대문자
# str.substr .... 문자열 자르기 등등
df2 = df['Name'].str.replace('.', '_').str.replace(',', '_').str.upper()


# 컬럼을 새로 추가하기
korean = np.random.randint(60, 100, 3)
math = np.random.randint(60, 100, 3)
english = np.random.randint(60, 100, 3)

# 새로운 컬럼명을 선택하고 값을 대입. 단일값은 모든 행에 단일값 적용. 리스트(ndArray)는 각 인덱스에 맞는 값으로 세팅
df['국어'] = korean
df['수학'] = math
df['영어'] = english


df['총합'] = 100
df2 = df.loc[:, '국어' : '영어']

# 특정 행렬이나 컬럼을 대상으로 연산을 수행할 수 있음
# sum, mean, max, min ...
df["총합"] = df2.sum(axis=1) # ndArray처럼 axis를 정해 세로, 가로 방향으로 연산 수행 가능
df

# 컬럼/인덱스 삭제 
df.drop('수학', axis = 1, inplace=True) # 수학이라는 이름의 컬럼을 삭제
df.drop(['국어', '영어'], axis=1, inplace=True) # 국어,영어 두개의 컬럼을 삭제

```
