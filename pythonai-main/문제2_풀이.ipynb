```python
import pandas as pd

df = pd.read_csv('students.csv')
```

```python
# 문제1. 학생들의 총점, 평균 구하기
# v1 

kor = df['국어']
eng = df['영어']
math = df['수학']


total_list = []
avg_list = []
for i in range(0, kor.count()) :
  total = kor[i] + eng[i] + math[i]
  avg = total / 3

  total_list.append(total)
  avg_list.append(avg)

total_list
avg_list

df = df.assign(총점 = total_list, 평균 = avg_list)
df

```

```python
# v2
## 시리즈 연산(벡터 연산)

li1 = [1,2,3]
li2 = [4,5,6]

li1 + li2

sr1 = pd.Series([1,2,3])
sr2 = pd.Series([4,5,6])

sr1 + sr2
sr1 * sr2

df['총점2'] = df['국어'] + df['영어'] + df['수학']
df['평균2'] = df['총점2'] / 3

df
```

```python
# v3 
# 집계함수
# sum -> 합계
df['총점3'] = df[['국어', '영어', '수학']].sum(axis=1)
df['평균3'] = df[['국어', '영어', '수학']].mean(axis=1)
# mean -> 평균
# max -> 최대값
# min -> 최소값
df
```

```python
# 문제2. 학생들의 평균이 높은 순으로 정렬. 만약 평균이 같다면 학번으로 오름차순
df2 = df.drop(['총점2', '총점3', '평균2', '평균3'], axis=1)

df2

df2.sort_values(['평균', '번호'], ascending=[False, True])
```

```python
# 문제3. 학생들의 평균으로 석차를 내어 석차 순으로 오름차순
df2 = df2.sort_values('평균', ascending=False)
df2
rank = range(1, 101)
df2['석차'] = rank
df2.sort_index()
```
