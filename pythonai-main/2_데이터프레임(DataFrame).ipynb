```python
import pandas as pd
# 2. 데이터 프레임 - 2차원 배열. 시리즈의 집합. 행렬(표)
# 2.1 데이터 프레임 생성
data = {
    "학번" : [1, 2, 3],
    "이름" : ["홍길동", "황진이", "이순신"],
    "학교" : ["서울대", "연세대", "고려대"], 
}

df = pd.DataFrame(data)
df
```

```python
# 2.2 데이터 프레임 index 설정
df2 = pd.DataFrame(data, index=['1번', '2번', '3번'])

df2
```

```python
# 2.3 데이터 프레임 column 설정
df3 = pd.DataFrame(data, index=['1번', '2번', '3번'], columns=['학번', '이름'])

df3
```

```python
# 2.4 데이터 프레임 컬럼명 바꾸기

## df의 정보 -> 컬럼
df.columns = ['No', 'Name', 'School']

## df의 정보 -> 인덱스
df.index = ['1번', '2번', '3번']

df
```
