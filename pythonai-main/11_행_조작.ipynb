```python
import pandas as pd

data = {
    "이름" : ["홍길동", "이순신", "황진이"],
    "나이" : [22, 33, 35]
}

df2 = pd.DataFrame(data)

df2['연령대'] = df2['나이'] // 10 * 10

df2
```

```python
# 11.1 행추가
df2['새컬럼'] = 'a'
df2['새컬럼2'] = ['a', 'b', 'c']

df2.drop(['새컬럼', '새컬럼2'], axis=1, inplace=True)
```

```python
df2.loc[3] = 'a'
df2.loc[4] = ['유관순', 25, 20]

df2
```

```python
# 11.2 행 삭제
df2.drop(3, axis=0, inplace=True)
```

```python
# 11.3 행 여러 개 삭제

df2.drop([1, 4], axis=0, inplace=True)
```

```python
df2
```

```python
# 11. 4 조건 삭제
df2.loc[3] = ['을지문덕', 34, 30]
df2.loc[4] = ['강감찬', 54, 50]
df2.loc[5] = ['유관순', 24, 20]
df2
## 조건에 맞는 행을 찾는다.
df2.loc[df2['연령대'] == 30].index
## 해당 행들의 index로 drop을 한다.

df2.drop(df2.loc[df2['연령대'] == 30].index, axis=0)
```

```python
# 11.5 인덱스 수정
df2.index = [0,1,2,3,4]

df2.index.values[0] = 100
df2
```

```python
# 11.6 원하는 행렬 데이터를 접근해서 수정 -> 데이터 선택의 연장

df2.loc[2, ['나이', '연령대']] = [44, 40]

df2
```
