```python
# 4. 파일 저장 및 열기 -

import pandas as pd

data = {
    "학번" : [1, 2, 3],
    "이름" : ["홍길동", "황진이", "이순신"],
    "학교" : ["서울대", "연세대", "고려대"], 
}

df = pd.DataFrame(data)
df
```

```python
df.columns = ['No', 'Name', 'School']

df.index = df['No']

df
```

```python
# csv, excel, json, txt
# 4.1 CSV로 데이터 프레임 저장

df.to_csv('myData1.csv') # -> 한글 깨짐

df.to_csv('myData2.csv', encoding='utf-8-sig')
```

```python
# 4.2 CSV 파일 읽어오기

df2 = pd.read_csv('/content/myData2.csv', index_col=0) # 인덱스 정보를 가져오지 않는 옵션션

df2
```

```python
# 4.3 CSV 인덱스 없이 저장하기
df.to_csv('myData3.csv', index=False, encoding='utf-8-sig')
```

```python
# 4.4 헤더 없이 저장
df.to_csv('myData4.csv', index=False, header=False, encoding='utf-8-sig')
```

```python
# 4.5 헤더 없는 파일 읽기
df3 = pd.read_csv('/content/myData4.csv', names=['학번', '이름', '학교'] )

df3
```
