```python
import pandas as pd

df = pd.read_excel('/content/거래내역조회.xlsx', skiprows=3)
```

```python
# 10. 정렬
# 10.1 시리즈 정렬 - 숫자
df['찾으신금액'].sort_values(ascending=True) # 오름차순 (ascending)
df['찾으신금액'].sort_values(ascending=False) # 내림차순

# 10.2 시리즈 정렬 - 문자
df['취급점'].sort_values()
df['취급점'].sort_values(ascending=False)

```

```python
# 10.3 데이터프레임 정렬
# 정렬할 기준 정하기
df.sort_values('찾으신금액') # 기본 오름차순
df.sort_values('찾으신금액', ascending=False) # 내림차순

# 10.4 기준이 여러개일 경우
df.sort_values(['취급점', '찾으신금액'], ascending=False)

# 10.5 기준이 여러개일 경우 오름차순, 내림차순 따로 정하기
df.sort_values(['취급점', '찾으신금액', '기재내용'], ascending=[True, False, False])

# 10.6 index로 정렬
df.sort_index()
df.sort_index(ascending=False)
```
