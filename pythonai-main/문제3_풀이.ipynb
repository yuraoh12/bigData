```python
# 문제1. 학생 데이터를 국어 성적으로 내림차순 정렬한 뒤 위에서 11개, 아래에서 7개 삭제
import pandas as pd

df = pd.read_csv('/content/students.csv')
df
```

```python
# 1. 국어로 내림차순

df2 = df.sort_values('국어', ascending=False)

# df2.drop(range(0, 11), axis=0)
# df2.loc[:11]

df2.index = df2['이름']
df2
#df2.loc['이민운'] # 인덱스 컬럼 값

df2.iloc[0:11] # 리스트 인덱스

df2.drop('이름', axis=1, inplace=True)
```

```python
idx1 = df2.iloc[0:11].index
idx2 = df2.iloc[99: 92: -1].index

df2.drop(idx1, axis=0, inplace=True)
df2.drop(idx2, axis=0, inplace=True)
```
